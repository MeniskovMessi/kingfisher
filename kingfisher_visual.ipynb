```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\menis\Desktop\kingfisher\kingfisher_data.xls')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=True, color='blue')
plt.title("Распределение цен")
plt.xlabel("Цена (тенге)")
plt.ylabel("Частота")
plt.show()
```

<h4>Гистограмма и распределение цен (в тенге) на изображении показывают сильно скошенное распределение с концентрацией значений вблизи нуля. Большинство цен сосредоточены на нижнем уровне шкалы, с резким снижением частоты при увеличении цен. Это говорит о том, что большинство товаров или данных имеют относительно низкие цены, в то время как только несколько позиций имеют значительно более высокие цены, что приводит к длинному правому хвосту в распределении.</h4>

```python
plt.figure(figsize=(8, 6))
category_counts = df['category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values, color='green')
plt.title("Количество товаров по категориям")
plt.xlabel("Категория")
plt.ylabel("Количество продуктов")
plt.xticks(rotation=45)
plt.show()
```

<h4>На гистограмме отображено количество товаров по категориям. Из графика видно, что категория "Бакалея" занимает лидирующую позицию с более чем 175 товарами, за ней следует категория "Антипасти" с чуть меньшим количеством. Другие категории, такие как "Рыба", "Снэки, сладости" и "Морепродукты", также имеют достаточно высокое количество товаров, но значительно меньшее по сравнению с лидерами. В то время как категории "Икра", "Масло", "Мясо" и "Сыры" имеют наименьшее количество товаров, что может свидетельствовать о меньшем разнообразии или ограниченном ассортименте в этих группах.</h4>

```python
plt.figure(figsize=(8, 6))
avg_rating_category = df.groupby('category')['rating'].mean().sort_values(ascending=False)
sns.barplot(x=avg_rating_category.index, y=avg_rating_category.values)
plt.title("Средний рейтинг по категориям")
plt.xlabel("Категория")
plt.ylabel("Средний балл")
plt.xticks(rotation=45)
plt.show()
```

<h4>На графике представлен средний рейтинг по категориям товаров. Видно, что категории "Икра", "Масло", "Сыры", "Морепродукты" и "Рыба" имеют высокие значения среднего рейтинга, приближаясь к 5 баллам. Это указывает на то, что эти категории товаров получают в основном положительные оценки. В то же время категории "Антипасти", "Бакалея", "Мясо" и "Снэки, сладости" не имеют рейтинги, что может свидетельствовать о более низком уровне увлеченности в оценке товаров в этих категориях.</h4>

```python
plt.figure(figsize=(8, 6))
avg_rating_category = df.groupby('subcategory')['rating'].mean().sort_values(ascending=False)
sns.barplot(x=avg_rating_category.index, y=avg_rating_category.values)
plt.title("Средний рейтинг по подкатегориям")
plt.xlabel("Податегория")
plt.ylabel("Средний балл")
plt.xticks(rotation=90)
plt.show()
```

<h4>На графике представлен средний рейтинг по подкатегориям товаров. Видно, что некоторые подкатегории имеют высокий средний рейтинг, близкий к 5 баллам. Например, подкатегории, такие как "Гребешок", "Замороженная рыба" и "Испанские сыры" получают максимальные оценки. Однако есть и несколько подкатегорий, которые имеют значительно более низкие оценки, такие как "Кальмары", "Охлажденная рыба", что указывает на низкую удовлетворенность, большая часть товаров не имеют в этих подкатегориях. В целом, график демонстрирует, что большинство подкатегорий, которые были оцененны клиентами имеют положительные рейтинги, но некоторые из них явно отстают.</h4>

```python
plt.figure(figsize=(10, 8))
subcategory_counts = df['subcategory'].value_counts()
sns.barplot(x=subcategory_counts.index, y=subcategory_counts.values, color='purple')
plt.title("Количество товаров по подкатегориям")
plt.xlabel("Подкатегория")
plt.ylabel("Количество продуктов")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

<h4>На графике представлено количество товаров по подкатегориям. Явным лидером является подкатегория "Антипасти Le Bonta", которая имеет наибольшее количество товаров, почти 70. За ней следуют такие подкатегории, как "Сладости", "Замороженная рыба" и "Черная икра". Эти подкатегории также имеют относительно большое количество товаров. В то же время, в нижней части графика представлены подкатегории с меньшим количеством товаров, такие как "Кофе", "Вонголе" и "Каперсы". Это может свидетельствовать о меньшем ассортименте в этих подкатегориях. В целом, график демонстрирует значительное разнообразие товаров в некоторых категориях и их концентрацию в других.</h4>

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='subcategory', y='price', data=df)

plt.title("Распределение цен по подкатегориям")
plt.xlabel("Подкатегория")
plt.ylabel("Цена (тенге)")

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

<h4>На графике показано распределение цен по подкатегориям товаров с помощью боксплотов. Из графика видно, что большинство подкатегорий имеют относительно низкие цены, так как медианные значения находятся в пределах 10,000-30,000 тенге. Однако есть несколько подкатегорий, например "Черная икра" и "Хамон", где наблюдаются значительно более высокие цены, что свидетельствует о присутствии дорогих товаров в этих категориях. В целом, график показывает значительное разнообразие цен по различным подкатегориям товаров.</h4>
